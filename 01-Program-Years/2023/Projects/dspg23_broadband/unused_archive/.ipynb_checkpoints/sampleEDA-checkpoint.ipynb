# Geocoding trial 1

```python
import geopy
import geopandas
```

```python
bk_lat_lon = pd.read_csv("data/BLACKKNIGHT_ANALYSIS_SUBSET.csv", low_memory=False,
                         usecols = [
                             'property_address_source_flag', 'property_full_street_address', 
                             'property_city_name', 'property_state', 'property_zip_code', 'property_zip4', 
                             'property_unit_type', 'property_unit_number', 'property_house_number', 
                             'property_street_direction_left', 'property_street_name', 
                             'property_street_suffix', 'property_street_direction_right', 
                             'property_address_carrier_route', 'property_address_latitiude', 
                             'property_address_longitude', 'property_address_census_tract'])
```

```python
bk_lat_lon.head()
```

```python
bk_lat_lon['address_line'] = bk_lat_lon.apply(lambda x: str(x["property_full_street_address"]) + ", " +
                                               str(x["property_city_name"]) + ", " + str(x['property_state']), axis = 1)
```

```python
bk_lat_lon.head()
```

```python
bk_lat_lon[bk_lat_lon["property_state"].isna()]
```

```python
bk_lat_lon.shape
```

```python
bk_lat_lon.shape
```

```python
from geopy.geocoders import Photon
geolocator = Photon(user_agent="dspg_ers_eda")
location = geolocator.geocode("1810 S Lakeport Street, Sioux City, IA")
print(location.latitude, location.longitude)
```

```python
location = geolocator.geocode("1810 Lakeport Street, 51106, Sioux City, Iowa, United States")
print(location.latitude, location.longitude)
```

```python
location = geolocator.reverse((location.latitude, location.longitude))
print(location.address)
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="dspg_ers_eda")
location = geolocator.geocode("1810 S Lakeport Street, Sioux City, IA", timeout=None)
print(location.latitude, location.longitude)
```

```python
from geopy.geocoders import Bing
BingMap_api = ""
geolocator = Bing(BingMap_api, user_agent = "dspg_ers_eda")
location = geolocator.geocode("1810 S Lakeport Street, Sioux City, IA")
print(location.latitude, location.longitude)
```

```python
from geopy.geocoders import Bing
geolocator = Bing(BingMap_api, user_agent = "dspg_ers_eda")
location = geolocator.geocode("1818 S Lakeport Street, Sioux City, IA")
print(location.latitude, location.longitude)
```
